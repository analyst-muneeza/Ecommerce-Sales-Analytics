DATA MERGING


```python
# Merge Products with Product Category

products = pd.merge(
    products,
    product_category,
    on="product_category_name",
    how="left"
)

print(products.head())
print(products.shape)
```

```python
# Merge Orders with Customers

orders_customers = pd.merge(
    orders,
    customers,
    on="customer_id",
    how="left"
)

print(orders_customers.head())
print(orders_customers.shape)
```

```python
# Merge Order Items

orders_customers_items = pd.merge(
    orders_customers,
    order_items,
    on="order_id",
    how="left"
)

print(orders_customers_items.head())
print(orders_customers_items.shape)
```

```python
# Merge Products

orders_products = pd.merge(
    orders_customers_items,
    products,
    on="product_id",
    how="left"
)

print(orders_products.head())
print(orders_products.shape)
```

```python
# Merge Payments

orders_payments = pd.merge(
    orders_products,
    payments,
    on="order_id",
    how="left"
)

print(orders_payments.head())
print(orders_payments.shape)
```

```python
# Merge Reviews

orders_reviews = pd.merge(
    orders_payments,
    reviews,
    on="order_id",
    how="left"
)

print(orders_reviews.head())
print(orders_reviews.shape)
```

```python
# Merge Sellers

final_data = pd.merge(
    orders_reviews,
    sellers,
    on="seller_id",
    how="left"
)

print(final_data.head())
print(final_data.shape)
```

```python
# Final Dataset Information

print(final_data.info())

print("\n")

print(final_data.shape)

print("\n")

print(final_data.columns)
```

```python
# Export Final Dataset

final_data.to_csv("Ecommerce_Final_Dataset.csv", index=False)

print("Final Dataset Exported Successfully")
```

EXPLORATORY DATA ANALYSIS (EDA)

```python
# Final Dataset Overview

print(final_data.shape)

print("\n")

print(final_data.head())

print("\n")

print(final_data.info())
```

```python
# Total Revenue

total_revenue = final_data["payment_value"].sum()

print("Total Revenue :", total_revenue)
```

```python
# Total Orders

total_orders = final_data["order_id"].nunique()

print("Total Orders :", total_orders)
```

```python
# Total Customers

total_customers = final_data["customer_unique_id"].nunique()

print("Total Customers :", total_customers)
```

```python
# Total Products Sold

total_products = final_data["product_id"].nunique()

print("Total Products :", total_products)
```

```python
# Monthly Sales Trend

monthly_sales = final_data.groupby("Order Month")["payment_value"].sum()

print(monthly_sales)

monthly_sales.plot(kind="bar", figsize=(10,5))

plt.title("Monthly Sales")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.show()
```

Insight

- Revenue fluctuates across the year, indicating seasonal purchasing behavior.
- May generated the highest revenue, while September and December recorded comparatively lower sales.

```python
# Order Status

status = final_data["order_status"].value_counts()

print(status)

status.plot(kind="bar", figsize=(8,5))

plt.title("Order Status")

plt.xlabel("Status")

plt.ylabel("Count")

plt.show()
```

Insight

- Most orders were successfully delivered, reflecting an efficient fulfillment process.
- Cancelled and unavailable orders account for only a small percentage of total orders.

```python
# Payment Type

payment = final_data["payment_type"].value_counts()

print(payment)

payment.plot(kind="bar", figsize=(8,5))

plt.title("Payment Type")

plt.xlabel("Payment Type")

plt.ylabel("Count")

plt.show()
```

Insight

- Credit cards are the most preferred payment method among customers.
- Alternative payment methods such as debit card and vouchers have significantly lower usage.

```python
# Top Customer States

top_states = final_data["customer_state"].value_counts().head(10)

print(top_states)

top_states.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Customer States")

plt.xlabel("State")

plt.ylabel("Orders")

plt.show()
```

Insight

- São Paulo (SP) contributes the highest number of customers.
- Marketing and expansion strategies can target other states with lower customer penetration.

```python
# Top Product Categories

top_category = final_data["product_category_name_english"].value_counts().head(10)

print(top_category)

top_category.plot(kind="bar", figsize=(10,5))

plt.title("Top Product Categories")

plt.xlabel("Category")

plt.ylabel("Count")

plt.show()
```

Insight

- Bed, Bath & Table is the best-selling product category.
- Home-related categories dominate customer purchases, indicating strong demand in this segment.

```python
# Delivery Status

delivery = final_data["Delivery Status"].value_counts()

print(delivery)

delivery.plot(kind="bar", figsize=(8,5))

plt.title("Delivery Status")

plt.xlabel("Status")

plt.ylabel("Orders")

plt.show()
```

Insight

- Most orders were delivered on time.
- Only a small proportion of deliveries were delayed, indicating strong logistics performance.

```python
# Review Score

review = final_data["review_score"].value_counts().sort_index()

print(review)

review.plot(kind="bar", figsize=(8,5))

plt.title("Review Score")

plt.xlabel("Rating")

plt.ylabel("Count")

plt.show()
```

Insight

- Most customers gave 5-star ratings, reflecting high customer satisfaction.
- Very few customers provided low ratings, suggesting limited service or product issues.

```python
# Revenue by Payment Type

revenue = final_data.groupby("payment_type")["payment_value"].sum()

print(revenue)

revenue.plot(kind="bar", figsize=(8,5))

plt.title("Revenue by Payment Type")

plt.xlabel("Payment Type")

plt.ylabel("Revenue")

plt.show()
```

Insight

- Credit card transactions contribute the highest share of total revenue.
- Other payment methods generate comparatively lower revenue.
